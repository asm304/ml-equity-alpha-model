# 02 — Feature Engineering
Build momentum/volatility/beta features and cache the feature matrix.

```python
import pandas as pd
monthly_stock_data = pd.read_parquet('../data/prices_monthly.parquet')
monthly_stock_data.tail()
```

```python
returns = monthly_stock_data.pct_change()
returns
```

```python
mom_feature = pd.concat({
    'mom_1m': returns.rolling(1).sum(),
    'mom_3m': returns.rolling(3).sum(),
    'mom_6m': returns.rolling(6).sum(),
    'mom_12m': returns.rolling(12).sum(),
    'rev_1m': -returns.rolling(1).sum(),
}, axis=1)
mom_feature.head()
```

```python
vol_6m = returns.rolling(6).std()
idvol_6m = (returns.sub(returns.mean(axis=1), axis=0)).rolling(6).std()
vol_feature = pd.concat({'vol_6m': vol_6m, 'idvol_6m': idvol_6m}, axis=1)
vol_feature.tail()
```

```python
assert 'SPY' in returns.columns, 'SPY missing from columns'
mr = returns['SPY']
mr
```

```python
cov = returns.rolling(24).cov(mr)
var = mr.rolling(24).var()
beta_24m = cov.div(var,axis=0)
beta_24m
```

```python
beta_24m = beta_24m.drop(columns=['SPY'],errors='ignore')
beta_24m
```

```python
beta_24m.columns = [f"beta_24m__{c}" for c in beta_24m.columns]
beta_24m.tail()
```

```python
X_raw = pd.concat([mom_feature,vol_feature],axis=1)
X_raw.tail()
```

```python
X_raw.columns = ["__".join(map(str, c)) if isinstance(c, tuple) else str(c) for c in X_raw.columns]
```

```python
def cs_zscore(df):
    return df.apply(lambda row: (row - row.mean()) / (row.std(ddof=1) + 1e-8), axis=1)

X_z = cs_zscore(X_raw)
X = pd.concat([X_z, beta_24m], axis=1).sort_index()
X.tail()
```

```python
X.to_parquet('../data/features_monthly.parquet')
X.tail()
```
